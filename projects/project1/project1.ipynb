```python
import folium

# Sample DataFrame
data = {
    'Latitude': [37.7749, 40.7128, 34.0522],
    'Longitude': [-122.4194, -74.0060, -118.2437],
    'City': ['San Francisco', 'New York', 'Los Angeles']
}
df = pd.DataFrame(data)

# Create a map
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)  # Centered in the US

# Add points
for _, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['City']).add_to(m)

# Display the map
m.save("map.html")  # Save as an HTML file
m
```
